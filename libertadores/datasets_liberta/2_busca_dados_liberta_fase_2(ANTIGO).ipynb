```python
# 2_busca_dados_liberta_fase_2.ipynb - Fluxo Corrigido e Organizado

# ======================================
# 📂 1. Carregar Dados da Fase 1
# ======================================

import pandas as pd
import json

# Caminho do arquivo Excel da Fase 1
caminho_classificacao_fase_1 = "classificacao_por_grupo_fase_1.xlsx"

# Carregar todas as abas do Excel
xls = pd.read_excel(caminho_classificacao_fase_1, sheet_name=None)

# Concatenar todas as abas em um único DataFrame
df_resultado_fase_1 = pd.concat(xls.values(), ignore_index=True)

# Visualizar para confirmar
display(df_resultado_fase_1)
print(df_resultado_fase_1['Grupo'].value_counts())

# ==========================================
# 📂 2. Gerar Classificação Inicial da Fase 2
# ==========================================

def gerar_classificados_fase2(df_resultado_fase_1):
    df_ordenado = df_resultado_fase_1.copy()

    df_ordenado["Ordem"] = df_ordenado.groupby("Grupo")["Pontos"].rank(method="first", ascending=False)
    df_ordenado = df_ordenado.sort_values(by=["Grupo", "Ordem"])

    classificados = df_ordenado.groupby("Grupo").head(2).copy()
    classificados["Posicao"] = classificados.groupby("Grupo").cumcount() + 1
    classificados["Letra_Grupo"] = classificados["Grupo"].str.extract(r"Grupo (\w)")[0]
    classificados["Chave"] = classificados["Posicao"].astype(str) + "_" + classificados["Letra_Grupo"]

    grupos_fase2 = {
        "Grupo I": ["1_A", "2_B", "1_C", "2_D"],
        "Grupo J": ["1_B", "2_A", "1_D", "2_C"],
        "Grupo K": ["1_E", "2_F", "1_G", "2_H"],
        "Grupo L": ["1_F", "2_E", "1_H", "2_G"],
    }

    dados_fase2 = []
    for grupo_fase2, chaves in grupos_fase2.items():
        for chave in chaves:
            if chave in classificados["Chave"].values:
                time_info = classificados[classificados["Chave"] == chave].iloc[0]
                dados_fase2.append({
                    "Grupo": grupo_fase2,
                    "Nome do Time": time_info["Nome do Time"],
                    "Pontos": 0,
                    "Vitorias": 0,
                    "Empates": 0,
                    "Derrotas": 0,
                    "Total Cartola": float(time_info["Total Cartola"]),
                    "Cartola Sofrido": 0,
                    "Saldo Cartola": 0,
                    "Posicao": int(time_info["Posicao"]),
                    "Letra_Grupo": time_info["Letra_Grupo"],
                    "Chave": chave
                })

    df_fase2 = pd.DataFrame(dados_fase2)
    return df_fase2

# Gerar df_fase2
df_fase2 = gerar_classificados_fase2(df_resultado_fase_1)

# Conferir
display(df_fase2)

# ========================================
# 📂 3. Gerar Confrontos da Fase 2
# ========================================

def gerar_confrontos_fase2(df_fase2):
    confrontos = []

    grupos_fase2 = {
        "Grupo I": ["1_A", "2_B", "1_C", "2_D"],
        "Grupo J": ["1_B", "2_A", "1_D", "2_C"],
        "Grupo K": ["1_E", "2_F", "1_G", "2_H"],
        "Grupo L": ["1_F", "2_E", "1_H", "2_G"],
    }

    rodadas_por_grupo = {
        1: [(1,2), (3,4)],
        2: [(2,3), (4,1)],
        3: [(1,3), (4,2)],
        4: [(2,1), (4,3)],
        5: [(3,2), (1,4)],
        6: [(3,1), (2,4)],
    }

    for rodada, partidas in rodadas_por_grupo.items():
        for grupo, chaves in grupos_fase2.items():
            for mandante_idx, visitante_idx in partidas:
                mandante_chave = chaves[mandante_idx - 1]
                visitante_chave = chaves[visitante_idx - 1]

                mandante_nome = df_fase2[df_fase2['Chave'] == mandante_chave]['Nome do Time'].values
                visitante_nome = df_fase2[df_fase2['Chave'] == visitante_chave]['Nome do Time'].values

                if mandante_nome.size > 0 and visitante_nome.size > 0:
                    confrontos.append({
                        "Rodada": rodada,
                        "Grupo": grupo,
                        "mandante": {"nome": mandante_nome[0], "pontos": None},
                        "visitante": {"nome": visitante_nome[0], "pontos": None}
                    })

    df_confrontos = pd.DataFrame(confrontos)
    return df_confrontos

# Gerar confrontos
df_confrontos_fase2 = gerar_confrontos_fase2(df_fase2)

# Conferir
display(df_confrontos_fase2)

# ==============================================
# 📂 4. Gerar Resultados Iniciais da Fase 2
# ==============================================

# Ajustar rodadas para começar na Rodada 7
df_resultados_fase2 = df_confrontos_fase2.copy()
df_resultados_fase2["rodada"] = df_resultados_fase2["Rodada"] + 6

# Ajustar colunas para exportação
df_resultados_fase2 = df_resultados_fase2[["rodada", "Grupo", "mandante", "visitante"]]
df_resultados_fase2.rename(columns={"Grupo": "grupo"}, inplace=True)

display(df_resultados_fase2)

# ===================================================
# 📂 5. Exportar os Arquivos para o Painel
# ===================================================

# Exportar classificacao_por_grupo_fase2.js
classificacao_fase2_js = {}
for grupo in df_fase2["Grupo"].unique():
    df_grupo = df_fase2[df_fase2["Grupo"] == grupo]
    classificacao_fase2_js[grupo] = []
    for _, row in df_grupo.iterrows():
        classificacao_fase2_js[grupo].append({
            "posicao": int(row["Posicao"]),
            "nome": row["Nome do Time"],
            "pontos": int(row["Pontos"]),
            "vitorias": int(row["Vitorias"]),
            "empates": int(row["Empates"]),
            "derrotas": int(row["Derrotas"]),
            "totalCartola": float(row["Total Cartola"]),
            "cartolaSofrido": float(row["Cartola Sofrido"]),
            "saldoCartola": float(row["Saldo Cartola"])
        })

with open("classificacao_por_grupo_fase2.js", "w", encoding="utf-8") as f:
    f.write("const classificacaoFase2 = ")
    f.write(json.dumps(classificacao_fase2_js, indent=2, ensure_ascii=False))
    f.write(";")

# Exportar confrontos_fase2_libertadores.js
conteudo_confrontos_js = "const confrontosFase2 = " + df_confrontos_fase2.to_json(orient="records", force_ascii=False, indent=2) + ";"

with open("confrontos_fase2_libertadores.js", "w", encoding="utf-8") as f:
    f.write(conteudo_confrontos_js)

# Exportar resultados_fase2.js
conteudo_resultados_js = "const resultadosFase2 = " + df_resultados_fase2.to_json(orient="records", force_ascii=False, indent=2) + ";"

with open("resultados_fase2.js", "w", encoding="utf-8") as f:
    f.write(conteudo_resultados_js)

print("\u2705 Todos os arquivos da Fase 2 gerados com sucesso!")
```

```python
from pathlib import Path
import nbformat as nbf


# ===============================================
# 📂 6. Calcular Classificação Atualizada Fase 2
# ===============================================
def calcular_classificacao_fase2(df_resultados, df_pontuacoes):
    estatisticas = {}

    for _, jogo in df_resultados.iterrows():
        grupo = jogo['grupo']
        mandante = jogo['mandante']['nome']
        visitante = jogo['visitante']['nome']
        pontos_mandante = jogo['mandante']['pontos']
        pontos_visitante = jogo['visitante']['pontos']

        if pontos_mandante is None or pontos_visitante is None:
            continue

        for time in [mandante, visitante]:
            if grupo not in estatisticas:
                estatisticas[grupo] = {}
            if time not in estatisticas[grupo]:
                estatisticas[grupo][time] = {
                    "Pontos": 0,
                    "Vitórias": 0,
                    "Empates": 0,
                    "Derrotas": 0,
                    "Total Cartola": 0,
                    "Cartola Sofrido": 0,
                }

        estatisticas[grupo][mandante]["Total Cartola"] += pontos_mandante
        estatisticas[grupo][mandante]["Cartola Sofrido"] += pontos_visitante

        estatisticas[grupo][visitante]["Total Cartola"] += pontos_visitante
        estatisticas[grupo][visitante]["Cartola Sofrido"] += pontos_mandante

        if pontos_mandante > pontos_visitante:
            estatisticas[grupo][mandante]["Pontos"] += 3
            estatisticas[grupo][mandante]["Vitórias"] += 1
            estatisticas[grupo][visitante]["Derrotas"] += 1
        elif pontos_mandante < pontos_visitante:
            estatisticas[grupo][visitante]["Pontos"] += 3
            estatisticas[grupo][visitante]["Vitórias"] += 1
            estatisticas[grupo][mandante]["Derrotas"] += 1
        else:
            estatisticas[grupo][mandante]["Pontos"] += 1
            estatisticas[grupo][visitante]["Pontos"] += 1
            estatisticas[grupo][mandante]["Empates"] += 1
            estatisticas[grupo][visitante]["Empates"] += 1

    dados = []
    for grupo, times in estatisticas.items():
        for nome, stats in times.items():
            dados.append({
                "Grupo": grupo,
                "Nome do Time": nome,
                "Pontos": stats["Pontos"],
                "Vitórias": stats["Vitórias"],
                "Empates": stats["Empates"],
                "Derrotas": stats["Derrotas"],
                "Total Cartola": stats["Total Cartola"],
                "Cartola Sofrido": stats["Cartola Sofrido"],
                "Saldo Cartola": stats["Total Cartola"] - stats["Cartola Sofrido"]
            })

    df_classificacao = pd.DataFrame(dados)
    df_classificacao.sort_values(
        by=["Grupo", "Pontos", "Vitórias", "Total Cartola", "Saldo Cartola"],
        ascending=[True, False, False, False, False],
        inplace=True
    )
    df_classificacao["Posicao"] = df_classificacao.groupby("Grupo").cumcount() + 1
    return df_classificacao

# Aplicar classificação com base em resultados existentes
df_classificacao_atualizada = calcular_classificacao_fase2(df_resultados_fase2, df_resultados_fase2)

# Visualizar
display(df_classificacao_atualizada)

# Exportar versão atualizada para JS
classificacao_fase2_js = {}
for grupo in df_classificacao_atualizada["Grupo"].unique():
    grupo_df = df_classificacao_atualizada[df_classificacao_atualizada["Grupo"] == grupo]
    classificacao_fase2_js[grupo] = []
    for _, row in grupo_df.iterrows():
        classificacao_fase2_js[grupo].append({
            "posicao": int(row["Posicao"]),
            "nome": row["Nome do Time"],
            "pontos": int(row["Pontos"]),
            "vitorias": int(row["Vitórias"]),
            "empates": int(row["Empates"]),
            "derrotas": int(row["Derrotas"]),
            "totalCartola": float(row["Total Cartola"]),
            "cartolaSofrido": float(row["Cartola Sofrido"]),
            "saldoCartola": float(row["Saldo Cartola"]),
        })

with open("classificacao_por_grupo_fase2.js", "w", encoding="utf-8") as f:
    f.write("const classificacaoFase2 = ")
    f.write(json.dumps(classificacao_fase2_js, indent=2, ensure_ascii=False))
    f.write(";")

```
